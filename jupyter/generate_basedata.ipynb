```python
import json
import torch
def load_jsonl(input_path) -> list:
    """
    Read list of objects from a JSON lines file.
    """
    data = []
    with open(input_path, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line.rstrip('\n|\r')))
    print('Loaded {} records from {}'.format(len(data), input_path))
    return data
```

```python
test = load_jsonl(f'dataset/bfly/train.jsonl')
```

```python
for corpus in ['train', 'dev', 'test']:
    dataset = load_jsonl(f'dataset/bfly/{corpus}.jsonl')
    dataset_list = [doc['article_original'] for doc in dataset if doc['article_original']]
    
    if corpus == 'dev'
        torch.save(dataset_list, f'dataset/bfly/bert_data/valid.pt')
    else:
        torch.save(dataset_list, f'dataset/bfly/bert_data/{corpus}.pt')
```
